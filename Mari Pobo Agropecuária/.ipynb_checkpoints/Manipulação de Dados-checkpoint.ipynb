```python
class Marcacoes:
    def __init__(self, nome, xmin, xmax, ymin, ymax):
        self.nome = nome
        self.xmin = xmin
        self.xmax = xmax
        self.ymin = ymin
        self.ymax = ymax
        self.predicaoAltura = 0
        self.predicaoLargura = 0
    
    def status(self):
        print('NAME: ' + self.nome.upper())
        print('xmin: ' + str(self.xmin))
        print('ymin: ' + str(self.ymin))
        print('xmax: ' + str(self.xmax))
        print('ymax: ' + str(self.ymax))
    
    def mostrarPredicao(self):
        print('NAME: ' + self.nome.upper())
        print('ALTURA: ' + str(self.predicaoAltura))
        print('LARGURA: ' + str(self.predicaoLargura))
```

```python
import os
import xml.etree.ElementTree as ET

listaDeAcerolaECartoes = []

caminhoMarcacaoXML = "./4_marcação_de_posição/"

for root,subFolder,files in os.walk(caminhoMarcacaoXML):
    for i,file in enumerate(files):
        
        tree =  ET.parse(caminhoMarcacaoXML+file)
        root = tree.getroot()
        
        for j,obj in enumerate(root.findall('object')):
            for propriedades in obj:
                if(propriedades.tag == "bndbox"):
                    
                    if(j > 5):
                        nome = "cartao_" + str(i+1)
                    else:
                        nome = "acerola_" + str(i+1) + "_" + str(j+1)

                    xmin = xmax = ymin = ymax = 0
                    
                    for posicao in propriedades:
                        if(posicao.tag == 'xmin'):
                            xmin = posicao.text
                        elif(posicao.tag == 'xmax'):
                            xmax = posicao.text
                        elif(posicao.tag == 'ymin'):
                            ymin = posicao.text
                        else:
                            ymax = posicao.text
                                
                    listaDeAcerolaECartoes.append(Marcacoes(nome, int(xmin), int(xmax), int(ymin), int(ymax)))
```

```python
listaMarcacaoAcerola = []
listaMarcacaoCartao = []

alturaRealCartao = 54
larguraRealCartao = 85

for i in  listaDeAcerolaECartoes:
    if(i.nome[0] == 'a'):
        listaMarcacaoAcerola.append(i)
    else:
        listaMarcacaoCartao.append(i)
        
for i,index in enumerate(listaMarcacaoCartao):

    cartaoAlturaPixel = index.ymax - index.ymin
    cartaoLarguraPixel = index.xmax - index.xmin
    
    for j in range(i*6,i*6 + 6):
        
        acerolaAlturaPixel = listaMarcacaoAcerola[j].ymax - listaMarcacaoAcerola[j].ymin
        acerolaLarguraPixel = listaMarcacaoAcerola[j].xmax - listaMarcacaoAcerola[j].xmin
        
        alturaRealAcerola = (acerolaAlturaPixel*alturaRealCartao)/cartaoAlturaPixel
        larguraRealAcerola = (acerolaLarguraPixel*larguraRealCartao)/cartaoLarguraPixel
        
        listaMarcacaoAcerola[j].predicaoAltura = alturaRealAcerola/10
        listaMarcacaoAcerola[j].predicaoLargura = larguraRealAcerola/10
```

```python
class AcerolaReal:
    def __init__(self, nome, altura, largura):
        self.nome = nome
        self.altura = altura
        self.largura = largura
    
    def status(self):
        print('====================================')
        print("NOME: " + self.nome)
        print("ALTURA: " + self.altura)
        print("LARGURA: " + self.largura)
```

```python
import os

caminhoMedidasReais = "./5_medidas_reais/"
medidasReais = []

for root,subFolder,allFiles in os.walk(caminhoMedidasReais):
    for fileName in allFiles:
        with open(caminhoMedidasReais+fileName,'r') as file:
            for i in range(6):
                nome = file.readline().splitlines() 
                labelAltura = file.readline().splitlines() 
                altura = file.readline().splitlines() 
                labelLargura = file.readline().splitlines() 
                largura = file.readline().splitlines()
                
                medidasReais.append(AcerolaReal(nome[0],altura[0],largura[0]))
```

```python
class JuncaoPredicaoReal:
    def __init__(self, nome, predicaoAltura, predicaoLargura, realAltura, realLargura):
        self.nome = nome
        self.predicaoAltura = predicaoAltura
        self.predicaoLargura = predicaoLargura
        self.realAltura = realAltura
        self.realLargura = realLargura
    
    def status(self):
        print('NAME: ' + self.nome.upper())
        print('ALTURA REAL: ' + str(self.realAltura))
        print('LARGURA REAL: ' + str(self.realLargura))
        print('ALTURA PREDITA: ' + str(self.predicaoAltura))
        print('LARGURA PREDITA: ' + str(self.predicaoLargura))
```

```python
juncaoFinal = []

for i in range(len(medidasReais)):
    
    predicaoAltura = listaMarcacaoAcerola[i].predicaoAltura
    predicaoLargura = listaMarcacaoAcerola[i].predicaoLargura
    
    realAltura = medidasReais[i].altura
    realLargura = medidasReais[i].largura
    
    juncaoFinal.append(JuncaoPredicaoReal("Acerola_" + str(i),
                                          round(predicaoAltura,2),
                                          round(predicaoLargura,2),
                                          float(realAltura),
                                          float(realLargura)))
maiorErro = 0
index = 0
for i in range(len(juncaoFinal)):
    if(i == 0):
        if(abs(juncaoFinal[i].realAltura - juncaoFinal[i].predicaoAltura) > abs(juncaoFinal[i].realLargura - juncaoFinal[i].predicaoLargura)):
            maiorErro = abs(juncaoFinal[i].realAltura - juncaoFinal[i].predicaoAltura)
            index = i
        else:
            maiorErro = abs(juncaoFinal[i].realLargura - juncaoFinal[i].predicaoLargura)
            index = i
    else:
        if(abs(juncaoFinal[i].realAltura - juncaoFinal[i].predicaoAltura) > abs(juncaoFinal[i].realLargura - juncaoFinal[i].predicaoLargura)):
            if(abs(juncaoFinal[i].realAltura - juncaoFinal[i].predicaoAltura) > maiorErro):
                maiorErro = abs(juncaoFinal[i].realAltura - juncaoFinal[i].predicaoAltura)
                index = i
            elif(abs(juncaoFinal[i].realLargura - juncaoFinal[i].predicaoLargura) > maiorErro):
                maiorErro = abs(juncaoFinal[i].realLargura - juncaoFinal[i].predicaoLargura)
                index = i           
print(juncaoFinal[index].status())
print(maiorErro)
```
