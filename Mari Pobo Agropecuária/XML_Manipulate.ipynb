```python
class Marcacoes:
    def __init__(self, nome, xmin, xmax, ymin, ymax):
        self.nome = nome
        self.xmin = xmin
        self.xmax = xmax
        self.ymin = ymin
        self.ymax = ymax
    
    def status(self):
        print('NAME: ' + self.nome.upper())
        print('xmin: ' + str(self.xmin))
        print('ymin: ' + str(self.ymin))
        print('xmax: ' + str(self.xmax))
        print('ymax: ' + str(self.ymax))
```
